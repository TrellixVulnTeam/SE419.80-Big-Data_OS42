```python
import numpy as np
import pandas as pd
```

```python
data_file = pd.read_csv('data_with_headers.csv')
#data_file = pd.read_excel('data_with_headers.xlsx')
```

```python
print(data_file[0:3])
```

```python
#parsing
time = data_file['time']
```

```python
sensors=data_file.ix[:,'s1':'s4']
#ignore deprecation error
print(sensors[0:6])
```

```python
# since time did not start at zero, do an offset
time = time - time[0]
print(time[0:3])
```

```python
# calculating for row averages
# secong parameter as 1 means row averages
avg_row = np.mean(sensors,1)
# second parameter as 0 means col averages
avg_col = np.mean(sensors,0)
print(avg_row[0:3])
print(avg_col)
```

```python
# exporting results
my_data = [time, sensors, avg_row]
result = pd.concat(my_data, axis=1)
# axis=1 means stacking columns one on top of the other left to right
print(result[0:3])
```

```python
#saving your result
result.to_csv('result.csv')
result.to_excel('result.xlsx')
result.to_html('result.htm')
result.to_clipboard()
```

```python
#plotting results
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# plot a figure
plt.figure(1)
#time vs sensors(s1 as red)
plt.plot(time, sensors['s1'],'r-')
#another plot:time vs average (average as dotted blue)
plt.plot(time,avg_row,'b.')
#adding a legend
plt.legend(['Sensor 1','Average'])
#adding x-label and y-label
plt.xlabel('Time (sec)')
plt.ylabel('Sensor Values')
#save as .PNG file
plt.savefig('myPlot.png')
```
